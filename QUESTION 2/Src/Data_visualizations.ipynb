```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
file_path = "cleaned_StudentsPerformance_data.csv"
df = pd.read_csv(file_path)
```

```python
# Create a new column for average score
df['average_score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# 1. Histogram - Distribution of Average Scores
plt.figure(figsize=(10, 6))
sns.histplot(df['average_score'], kde=True, bins=20, color='blue')
plt.title('Distribution of Average Test Scores')
plt.xlabel('Average Score')
plt.ylabel('Frequency')
plt.savefig("visualization1.png")
plt.show()


```

```python
# 2. Box Plot - Gender vs. Average Scores
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='average_score', data=df, palette='Set2')
plt.title('Gender vs. Average Test Scores')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.savefig("visualization2.png")
plt.show()

```

```python
# 3. Bar Chart - Parental Education Level vs. Average Scores
plt.figure(figsize=(12, 6))
education_avg_scores = df.groupby('parental level of education')['average_score'].mean().sort_values()
education_avg_scores.plot(kind='bar', color='purple')
plt.title('Parental Education Level vs. Average Test Scores')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.savefig("visualization3.png")
plt.show()

```

```python
# 4. Scatter Plot - Test Preparation Course vs. Average Scores
plt.figure(figsize=(10, 6))
sns.stripplot(x='test preparation course', y='average_score', data=df, jitter=True, palette='Set1')
plt.title('Test Preparation Course vs. Average Scores')
plt.xlabel('Test Preparation Course')
plt.ylabel('Average Score')
plt.savefig("visualization4.png")
plt.show()

```

```python
# 5. Pie Chart - Lunch Type Distribution
plt.figure(figsize=(8, 8))
df['lunch'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightblue', 'lightgreen'])
plt.title('Lunch Type Distribution')
plt.ylabel('')
plt.savefig("visualization5.png")
plt.show()
```
